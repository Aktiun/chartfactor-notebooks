```python
from chartfactor import *
```

```python
import pandas as pd
```

```python
titanic = pd.read_csv('https://chartfactor.com/resources/tutorials/titanic.csv')
```

```python
cf.studio("My titanic app")
```

```python
cf.source("titanic")\
.graph("Histogram")\
.metrics(
	cf.Metric("Age", "histogram").fixedBars(6)
)\
.limit(100)\
.set("color", cf.Color().palette(["#0095b7","#a0b774","#f4c658","#fe8b3e","#cf2f23","#756c56","#007896","#47a694","#f9a94b","#ff6b30","#e94d29","#005b76"]).theme({"name":"light","font":"black","background":"white"}))\
.title("Age histogram")\
.execute()
```
