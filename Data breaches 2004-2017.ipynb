```python
from chartfactor import *
import pandas as pd
```

```python
dataBreaches = pd.read_csv('https://aktiun-data.s3.amazonaws.com/public/Data_Breaches_EN_V2_2004_2017_20180220.csv')
```

```python
cf.studio("Data breaches study 2004-2017")
```

```python
cf.source("dataBreaches")\
.graph("Bars")\
.groupby(
	cf.Attribute("Sector").limit(10).sort("desc", cf.Metric("Records Lost", "sum"))
)\
.metrics(
	cf.Metric("Records Lost", "sum")
)\
.filters(
	cf.Filter('Year').label('Year').operation('IN').value([2017])
)\
.clientFilters(
	cf.Filter('Sector').label('Sector').operation('IN').value(["Web"])
)\
.set("color", cf.Color().palette(["#084594","#2171b5","#4292c6","#6baed6","#9ecae1","#c6dbef","#eff3ff"]).metric(cf.Metric("Records Lost", "sum")).range([{"min":13470000,"max":71315232,"color":"#eff3ff"},{"min":71315232,"max":131559719,"color":"#c6dbef"},{"min":131559719,"max":217330755,"color":"#9ecae1"},{"min":217330755,"max":242978430,"color":"#6baed6"},{"min":242978430,"max":844079083,"color":"#4292c6"},{"min":844079083,"max":1338578353,"color":"#2171b5"},{"min":1338578353,"max":5776239450,"color":"#ef1b0b"}]).theme({"name":"light","font":"black","background":"white"}))\
.set("legend", cf.Legend().position("right").width(150).height('95%').sort("none"))\
.set("grid", cf.Grid().top(10).right(15).bottom(35).left(65))\
.set("orientation", "horizontal")\
.set("xAxis", {"show":True,"lines":True})\
.set("yAxis", {"text":"out"})\
.set("dataZoom", False)\
.title("Records lost by the web sector in 2017")\
.execute()
```
