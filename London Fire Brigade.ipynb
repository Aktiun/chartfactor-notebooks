```python
from chartfactor import *
```

```python
import pandas as pd
```

```python
london_fire_calls = pd.read_csv('https://chartfactor.com/resources/tutorials/london_fire_brigade.csv', parse_dates=['date_of_call','timestamp_of_call'], infer_datetime_format=True)
```

```python
cf.studio("London Fire Brigade Calls")
```

```python
cf.source("london_fire_calls")\
.graph("Vector Map")\
.groupby(
	cf.Attribute("proper_case").limit(10000).sort("desc", cf.Metric())
)\
.metrics(
	cf.Metric("count")
)\
.set("color", cf.Color().palette(["#a50026","#d73027","#f46d43","#fdae61","#fee090"]).metric(cf.Metric("count")).theme({"name":"light","font":"black","background":"white"}).autoRange(True))\
.set("shape", "https://chartfactor.com/resources/london.json")\
.set("enableZoom", False)\
.set("min", 0)\
.set("zoom", 0.428105536714385)\
.set("center", [-0.016937899627567088,51.485514506540866])\
.set("legend", "right")\
.title("Heat Map with boroughs")\
.execute()
```
