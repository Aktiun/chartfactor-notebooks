```python
from chartfactor import *
import pandas as pd
```

```python
healthy_cities = pd.read_csv('https://chartfactor.com/resources/tutorials/healthy_lifestyle_city_2021.csv')
```

```python
cf.studio("Healthy Lifestyle Cities ")
```

```python
cf.source("healthy_cities")\
.graph("Scatter Plot")\
.groupby(
	cf.Attribute("City").limit(100).sort("desc", cf.Metric("Life expectancy(years) (Country)", "sum"))
)\
.metrics(
	cf.Metric("Happiness levels(Country)", "sum"),
	cf.Metric("Life expectancy(years) (Country)", "sum"),
	cf.Metric("Outdoor activities(City)", "sum")
)\
.filters(
	cf.Filter('Happiness levels(Country)').label('Happiness levels(Country)').operation('BETWEEN').value([7,7.8]),
	cf.Filter('Life expectancy(years) (Country)').label('Life expectancy(years) (Country)').operation('BETWEEN').value([80,83.2])
)\
.set("color", cf.Color().palette(["#0095b7","#a0b774","#f4c658","#fe8b3e","#cf2f23","#756c56","#007896","#47a694","#f9a94b","#ff6b30","#e94d29","#005b76"]).theme({"name":"light","font":"black","background":"white"}).match({"Fukuoka":"#0095b7","Tokyo":"#a0b774","Milan":"#f4c658","Zurich":"#fe8b3e","Geneva":"#cf2f23","Madrid":"#756c56","Barcelona":"#007896","Melbourne":"#47a694","Sydney":"#f9a94b","Tel Aviv":"#ff6b30","Stockholm":"#e94d29","Paris":"#005b76","Vancouver":"#0095b7","Toronto":"#a0b774","Seoul":"#f4c658","Amsterdam":"#fe8b3e","Vienna":"#cf2f23","Frankfurt":"#756c56","Berlin":"#007896","Dublin":"#47a694","Brussels":"#f9a94b","London":"#ff6b30","Helsinki":"#e94d29","Copenhagen":"#005b76","Los Angeles":"#0095b7","Washington, D.C.":"#a0b774","Chicago":"#f4c658","New York":"#fe8b3e","San Francisco":"#cf2f23","Boston":"#756c56","Mexico City":"#007896","Buenos Aires":"#47a694","Hong Kong":"#f9a94b","Taipei":"#ff6b30","Beijing":"#e94d29","Shanghai":"#005b76","Istanbul":"#0095b7","Bangkok":"#a0b774","Sao Paulo":"#f4c658","Cairo":"#fe8b3e","Moscow":"#cf2f23","Jakarta":"#756c56","Mumbai":"#007896","Johannesburg":"#47a694"}))\
.set("legend", cf.Legend().position("right").width(150).height('95%').sort("none"))\
.set("grid", cf.Grid().top(10).right(25).bottom(35).left(40))\
.title("Cities with top happiness levels and life expectancy")\
.execute()
```
